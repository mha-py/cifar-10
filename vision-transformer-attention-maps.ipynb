```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm, trange
import pickle as pkl

%load_ext autoreload
%autoreload 2
from utils import *
from layers import *
from transformer import *
```

```python
path = 'F:/$Daten/datasets/cifar10/cifar-10-batches-py/'
fn = 'data_batch_1'

with open(path+fn, 'rb') as f:
    dataset = pkl.load(f, encoding='bytes')

N = len(dataset[b'labels'])
```

```python
def showimg(im):
    plt.imshow(im, extent=(0,1,0,1))
    plt.axis('off')
    #plt.show()

def getimg(i):
    return dataset[b'data'][i].reshape((3, 32, 32)).transpose((1, 2, 0)) / 255
def getlabel(i):
    return dataset[b'labels'][i]
```

```python
showimg(getimg(1))
getlabel(1)
```

```python

def batchgen(bsize=64, start=500):
    ep = 0
    while True:
        inds = np.random.permutation(range(start, N))
        minibatches = [ inds[k*bsize:(k+1)*bsize] for k in range(len(inds)//bsize) ]
        for mb in minibatches:
            xs = []
            ys = []
            for i in mb:
                xs.append(getimg(i))
                ys.append(getlabel(i))
            yield np.array(xs), np.array(ys)
        print(f'========== EPOCH {ep} COMPLETED ==========')
        ep += 1
```

```python
bg = batchgen()
xs, ys = next(bg)
xs.shape
```

```python
def valloss():
    net.eval()
    xs = np.array([getimg(i) for i in range(200)])
    yt = np.array([getlabel(i) for i in range(200)])
    xs, yt = np2t(xs, yt)
    yt = yt.long()
    yp = net(xs)
    loss = F.nll_loss(F.log_softmax(yp, dim=1), yt).item()
    yp, yt = t2np(yp, yt)
    yp = yp.argmax(-1)
    acc = np.mean(yp==yt)
    return loss, acc
    
def loss():
    net.train()
    xs, yt = next(bg)
    xs, yt = np2t(xs, yt)
    yt = yt.long()
    yp = net(xs)
    return F.nll_loss(F.log_softmax(yp, dim=1), yt)
```

```python
net = Net(128, 1)
from torch_optimizer import Lookahead, Yogi
net.optim = Lookahead(Yogi(net.parameters(), lr=3e-3))
net.iters = 0
net.losses = []
net.vlosses = []
net.vaccs = []
bg = batchgen()

from torchsummary import summary
summary(net, (32, 32, 3))
```

```python
losses = []

for k in trange(999999):
    net.train()
    l = loss()
    l.backward()
    losses.append(l.item())
    net.optim.step()
    net.zero_grad()

    if len(losses) == 50:
        vl, vacc = valloss()
        net.vlosses.append((net.iters, vl))
        net.vaccs.append((net.iters, vacc))
        net.losses.append((net.iters, np.mean(losses)))
        losses = []

    if k % 50 == 0:
        plt.plot(*zip(*net.losses))
        plt.plot(*zip(*net.vlosses))
        plt.plot(*zip(*net.vaccs))
        plt.grid()
        plt.show()

    net.iters += 1
```
